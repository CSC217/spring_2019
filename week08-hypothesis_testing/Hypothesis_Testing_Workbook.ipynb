Now we can plot a critical value by using the 'fill between' function. The fill between function takes three arguments. The first is a list of X-values - in this case a list of X-values between the 0.1st percentile and the 20th percentile. The second is the bottom limit of the fill space - in this case 0. The third is the upper limit of the fill space, in this case the PDF of the list of X-values.

```python
#Plot a PDF Graph with Mean 100 and standard deviation of 5 (based on the survey population of 9)
dist = stats.norm(100, 15/np.sqrt(9))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.2), dist.ppf(0.2)],[0, dist.pdf(dist.ppf(0.2))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.0001), dist.ppf(0.2))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('IQ')
plt.ylabel('Probability Density Function')
fig.suptitle('Distribution of IQ Among the Population', fontsize=15, y=0.92)
```

Now let's try a variation of a problem we did in class. We now interview 35 people instead of 9 at CCNY, and find that they have a mean IQ of 105. At the 0.05 significance level, can we reject the null hypothesis that CCNY students have the same IQ as the general population?

We only care if they are greater than or equal to the general population, so this a **one-sided** test.

```python
dist = stats.norm(100, 15/np.sqrt(35))
dist.ppf(0.95)
```

The 95th percentile of our distribution is 104.17, meaning that the rejection region is any IQ that falls above it. Thus, at the 0.05 significance level, we can reject the null hypothesis that CCNY students have the same IQ as the general population.

What about at the 0.01 significance level?

```python
dist = stats.norm(100, 15/np.sqrt(35))
dist.ppf(0.99)
```

Because the threshold is now above 105, we fail to reject the null hypothesis that CCNY students have the same IQ as the general population.

```python
#One-sided test at the 95th percentile - note that 105 is in the rejection region
dist = stats.norm(100, 15/np.sqrt(35))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.95), dist.ppf(0.95)],[0, dist.pdf(dist.ppf(0.95))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.95), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('IQ')
plt.ylabel('Probability Density Function')
fig.suptitle('Distribution of IQ Among the Population', fontsize=15, y=0.92)
```

```python
#One-sided test at the 99th percentile - note that 105 is no longer in the rejection region
dist = stats.norm(100, 15/np.sqrt(35))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.99), dist.ppf(0.99)],[0, dist.pdf(dist.ppf(0.99))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.99), dist.ppf(0.9999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('IQ')
plt.ylabel('Probability Density Function')
fig.suptitle('Distribution of IQ Among the Population', fontsize=15, y=0.92)
```

### Fun? Diversion

Before we go onto our next example - a fun diversion. I obviously made up the example of the IQ of CCNY students. Originally I just posited a sample mean of 110, but wanted to make a made up sample set of nine, normally distributed, with a mean of 110 and standard deviation of 20.

We covered a few classes ago that you can generate random numbers from a distribution with the 'rvs' function. Here we'll generate 

```python
np.random
sample_dist = stats.norm(110, 20).rvs(9)
sample_dist
```

However, the *sample* doesn't have this same mean and standard deviation (even though it's close here). This is because even though it's *pulling* from a normal distribution, it doesn't mean that the sample necessarily itself has that same mean and standard deviation of the original distribution (though a *series* of sample means pulled from the distribution will have the same mean as the original distribution :))

```python
sample_dist.mean()
```

```python
sample_dist.std()
```

Below is a solution to get a sample dataset that has the exact size, mean, and standard deviation that you want (it's how I developed the one we used in class). This is beyont the scope of the class, but may be useful for those of you who are interested in how to simulate data.

     1) Create a dataset of your desired length pulling from a normal distribution with a mean of 0 and your desired standard deviation. Note that again, even though you specify a mean of 0, the true center of the sample isn't 0.
     2) Re-center the sample by subtracting each value in the sample by the sample mean, so that the sample actually has a mean of 0.
     3) Now re-scale the sample by multiplying each value in the sample by the desired standard deviation over the sample standard deviation, so that the sample now has the correct standard deviation (still with a mean of 0).
     4) Now re-center the sample by adding each value in the sample by the desired mean, so now the sample actually has the correct mean!

```python
np.random.seed(42)
desired_mean=110
desired_std=20
desired_length=9

samples = stats.norm(0, desired_std).rvs(desired_length)

actual_mean = np.mean(samples)
actual_std = np.std(samples)
print("Initial samples stats   : mean = {:.4f} stdv = {:.4f}".format(actual_mean, actual_std))

zero_mean_samples = samples - (actual_mean)

zero_mean_mean = np.mean(zero_mean_samples)
zero_mean_std = np.std(zero_mean_samples)
print("True zero samples stats : mean = {:.4f} stdv = {:.4f}".format(zero_mean_mean, zero_mean_std))

scaled_samples = zero_mean_samples * (desired_std/zero_mean_std)
scaled_mean = np.mean(scaled_samples)
scaled_std = np.std(scaled_samples)
print("Scaled samples stats    : mean = {:.4f} stdv = {:.4f}".format(scaled_mean, scaled_std))

final_samples = scaled_samples + desired_mean
final_mean = np.mean(final_samples)
final_std = np.std(final_samples)
print("Final samples stats     : mean = {:.4f} stdv = {:.4f}".format(final_mean, final_std))
print(final_samples)
```

## Estimation

The sample mean is simply the mean of the sample.

```python
final_samples_int = [int(np.round(i, 0)) for i in final_samples]
np.mean(final_samples_int)
```

Remember that we can manually find the variance of the sample...

```python
z = 0
for i in final_samples_int:
    z = z + np.power(i - np.mean(final_samples_int), 2)
print(z/len(final_samples_int))
```

Or find it with Numpy

```python
np.var(final_samples_int)
```

For the same variance we are dividing by the number of samples minus 1

```python
z = 0
for i in final_samples_int:
    z = z + np.power(i - np.mean(final_samples_int), 2)
print(z/(len(final_samples_int) - 1))
```

Or find it with Numpy

```python
np.var(final_samples_int, ddof=1)
```

```python
np.std(final_samples_int, ddof=1)
```

We can use this mean and standard deviation to build a T-Distribution, which, as discussed in class, is similar to a normal distribution with fatter tails.

Our T-distribution, for a sample of 9, will have a mean of 100 (our null hypothesis), along with a standard deviation equal to our sample standard deviation divided by the square root of our sample size.

```python
dist = stats.t(loc=100, scale=np.std(final_samples_int, ddof=1)/np.sqrt(9), df=8)
dist.mean()
```

```python
dist.ppf(0.025)
```

```python
dist.ppf(0.975)
```

```python
#Plot a PDF Graph with Mean 100 and standard deviation of 5 (based on the survey population of 9)
dist = stats.t(loc=100, scale=np.std(final_samples_int, ddof=1)/np.sqrt(9), df=8)
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.025), dist.ppf(0.025)],[0, dist.pdf(dist.ppf(0.025))], linestyle='dotted', color='#1f77b4')
plt.plot([dist.ppf(0.975), dist.ppf(0.975)],[0, dist.pdf(dist.ppf(0.025))], linestyle='dotted', color='#1f77b4')
plt.ylim(bottom=0)
plt.xlabel('IQ')
plt.ylabel('Probability Density Function')
fig.suptitle('T-Distribution of IQ Among the Population (8 degrees of Freedom)', fontsize=15, y=0.92)
```
